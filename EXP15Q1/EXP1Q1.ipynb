```python
# Create Rectangle class with attributes length and breadth and methods to find area and perimeter. Compare two Rectangle objects by their area.

class rectangle:
    def get(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l*self.b
    def perimeter(self):
        return 2*(self.l+self.b)
r1=rectangle()
r2=rectangle()
l1=int(input("Enter length of Rectangle1: "))
b1=int(input("Enter breadth of Rectangle1: "))
l2=int(input("Enter length of Rectangle2: "))
b2=int(input("Enter breadth of Rectangle2: "))
r1.get(l1,b1)
r2.get(l2,b2)
p=r1.area()
q=r2.area()
per1=r1.perimeter()
per2=r2.perimeter()
print("Area of Rectangle1:", p)
print("Perimeter of Rectangle1:",per1)
print("Area of Rectangle2:", q)
print("Perimeter of Rectangle2:",per2)
if p>q:
    print("Rectangle 1 is greater")
elif p<q:
    print("Rectangle 2 is greater")
else:
    print("Rectangles are equal")
```
